```python
for par in df.describe().T.index[:5]:
    mean = df.describe().T.loc[par, 'mean']
    median = df.describe().T.loc[par, '50%']
    mean_vs_median = round(abs(mean - median) / mean , 2)
    print("mean vs median for {} is {}".format(par, mean_vs_median))
```
